```python
import csv
import requests
from bs4 import BeautifulSoup
import time
import random

# --- CONFIGURATION ---
input_file = 'Amazon_Best_Seller_2021_June 2.csv'
output_file = 'resultats_titres.csv'

def get_amazon_title(asin):
    """Va chercher le titre sur Amazon sans utiliser Pandas"""
    url = f"https://www.amazon.com/dp/{asin}"
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36",
        "Accept-Language": "en-US,en;q=0.9"
    }
    
    try:
        response = requests.get(url, headers=headers, timeout=10)
        if response.status_code == 200:
            soup = BeautifulSoup(response.content, "html.parser")
            # On cherche le titre dans la page
            title_tag = soup.find("span", id="productTitle")
            if title_tag:
                return title_tag.get_text().strip()
    except:
        pass
    return "Titre non trouvé"

print("Démarrage de l'extraction...")

# On ouvre le fichier source et le fichier de destination
with open(input_file, 'r', encoding='utf-8', errors='replace') as f_in, \
     open(output_file, 'w', newline='', encoding='utf-8') as f_out:
    
    reader = csv.reader(f_in)
    writer = csv.writer(f_out)
    
    # Créer les colonnes du nouveau fichier
    writer.writerow(['ASIN', 'Titre Récupéré'])
    
    next(reader, None) # On saute la première ligne (en-têtes)
    
    for i, row in enumerate(reader):
        if not row: continue # Sécurité ligne vide
        
        asin = row[0] # La colonne 0 contient l'ASIN
        print(f"[{i+1}] Récupération de {asin}...", end="")
        
        titre = get_amazon_title(asin)
        writer.writerow([asin, titre])
        
        print(f" OK")
        time.sleep(1) # Petite pause pour éviter le blocage

print(f"Fini ! Ouvrez le fichier '{output_file}' pour voir les résultats.")
```

```python
import csv
import requests
from bs4 import BeautifulSoup
import time
import random

# --- CONFIGURATION ---
input_file = 'Amazon_Best_Seller_2021_June 2.csv'
output_file = 'resultats_par_lien.csv'

def get_title_from_url(url):
    """Récupère le titre depuis une URL Amazon donnée."""
    # En-têtes pour imiter un vrai navigateur (indispensable pour Amazon)
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36",
        "Accept-Language": "en-US,en;q=0.9"
    }
    
    try:
        response = requests.get(url, headers=headers, timeout=10)
        if response.status_code == 200:
            soup = BeautifulSoup(response.content, "html.parser")
            
            # 1. Essai sur la page "Offer Listing" (le format de vos liens)
            title = soup.find("h1")
            if title:
                return title.get_text().strip()
                
            # 2. Essai générique (balise ID souvent utilisée)
            title = soup.find("span", id="productTitle")
            if title:
                return title.get_text().strip()
                
            # 3. Dernier recours : le titre de la page HTML
            if soup.title:
                return soup.title.get_text().replace("Amazon.com: Buying Choices:", "").strip()

    except Exception as e:
        return f"Erreur : {e}"
        
    return "Titre non trouvé"

print("Démarrage du scraping via les liens...")

with open(input_file, 'r', encoding='utf-8', errors='replace') as f_in, \
     open(output_file, 'w', newline='', encoding='utf-8') as f_out:
    
    reader = csv.reader(f_in)
    writer = csv.writer(f_out)
    
    # Création des colonnes du fichier de sortie
    writer.writerow(['ASIN', 'Lien Original', 'Titre Extrait'])
    
    # Sauter l'en-tête du fichier source
    next(reader, None)
    
    for i, row in enumerate(reader):
        if not row: continue
        
        asin = row[0]
        # Le lien est dans la colonne 2 (index 2)
        # IMPORTANT : On remplace '&amp;' par '&' pour que le lien fonctionne
        raw_link = row[2]
        clean_link = raw_link.replace("&amp;", "&")
        
        print(f"[{i+1}] Scraping : {clean_link[:50]}...", end="")
        
        titre = get_title_from_url(clean_link)
        
        writer.writerow([asin, clean_link, titre])
        print(" OK")
        
        # Pause aléatoire entre 1 et 3 secondes pour éviter le blocage
        time.sleep(random.uniform(1, 3))

print(f"Extraction terminée ! Voir le fichier : {output_file}")
```

```python
import pandas as pd
import asyncio
import aiohttp
import random
from bs4 import BeautifulSoup

# ==========================
# CONFIG
# ==========================
INPUT_CSV = "Amazon_Best_Seller_2021_June 2.csv"
OUTPUT_CSV = "amazon_results.csv"

ASIN_COLUMN = "ASIN"   # <-- adapte si besoin
CHUNK_SIZE = 50
MAX_CONCURRENCY = 5

HEADERS_LIST = [
    {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)",
        "Accept-Language": "en-US,en;q=0.9"
    },
    {
        "User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10_15_7)",
        "Accept-Language": "en-US,en;q=0.9"
    }
]

# ==========================
# ASYNC FETCH
# ==========================
async def fetch_product(session, asin):
    url = f"https://www.amazon.com/dp/{asin}"
    headers = random.choice(HEADERS_LIST)

    try:
        async with session.get(url, headers=headers, timeout=15) as resp:
            if resp.status != 200:
                return {"asin": asin, "title": None, "status": resp.status}

            html = await resp.text()
            soup = BeautifulSoup(html, "html.parser")

            title_tag = soup.select_one("#productTitle")
            title = title_tag.text.strip() if title_tag else None

            return {
                "asin": asin,
                "title": title,
                "status": resp.status
            }

    except Exception as e:
        return {"asin": asin, "title": None, "status": "error"}

# ==========================
# PROCESS CHUNK
# ==========================
async def process_chunk(asins):
    connector = aiohttp.TCPConnector(limit=MAX_CONCURRENCY, ssl=False)
    async with aiohttp.ClientSession(connector=connector) as session:
        tasks = [fetch_product(session, asin) for asin in asins]
        return await asyncio.gather(*tasks)

# ==========================
# MAIN
# ==========================
async def main():
    df = pd.read_csv(INPUT_CSV)
    asins = df[ASIN_COLUMN].dropna().unique().tolist()

    print(f"🔹 Total ASIN: {len(asins)}")

    all_results = []

    for i in range(0, len(asins), CHUNK_SIZE):
        chunk = asins[i:i + CHUNK_SIZE]
        print(f"⏳ Chunk {i//CHUNK_SIZE + 1}")

        results = await process_chunk(chunk)
        all_results.extend(results)

        await asyncio.sleep(random.uniform(3, 6))  # pause intelligente

    out_df = pd.DataFrame(all_results)
    out_df.to_csv(OUTPUT_CSV, index=False)

    print("✅ Scraping terminé")

# ==========================
# RUN
# ==========================
if __name__ == "__main__":
    asyncio.run(main())
```

```python
import pandas as pd
import asyncio
import aiohttp
import random
from bs4 import BeautifulSoup

# ==========================
# CONFIGURATION
# ==========================
INPUT_CSV = "Amazon_Best_Seller_2021_June 2.csv"
OUTPUT_CSV = "amazon_output.csv"

ASIN_COLUMN = "ASIN"     # ⚠️ adapte si besoin
CHUNK_SIZE = 50
MAX_CONCURRENCY = 5

HEADERS = [
    {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)",
        "Accept-Language": "en-US,en;q=0.9"
    },
    {
        "User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10_15_7)",
        "Accept-Language": "en-US,en;q=0.9"
    }
]

# ==========================
# SCRAPE UN PRODUIT
# ==========================
async def fetch_product(session, asin):
    url = f"https://www.amazon.com/dp/{asin}"
    headers = random.choice(HEADERS)

    try:
        async with session.get(url, headers=headers, timeout=15) as resp:
            status = resp.status
            html = await resp.text()

            soup = BeautifulSoup(html, "html.parser")
            title_tag = soup.select_one("#productTitle")

            title = title_tag.text.strip() if title_tag else None

            return {
                "ASIN": asin,
                "title": title,
                "status": status
            }

    except Exception as e:
        return {
            "ASIN": asin,
            "title": None,
            "status": "error"
        }

# ==========================
# TRAITER UN CHUNK
# ==========================
async def process_chunk(asins):
    connector = aiohttp.TCPConnector(limit=MAX_CONCURRENCY, ssl=False)
    async with aiohttp.ClientSession(connector=connector) as session:
        tasks = [fetch_product(session, asin) for asin in asins]
        return await asyncio.gather(*tasks)

# ==========================
# MAIN
# ==========================
async def main():
    df = pd.read_csv(INPUT_CSV)

    asins = (
        df[ASIN_COLUMN]
        .dropna()
        .astype(str)
        .unique()
        .tolist()
    )

    print(f"🔹 Total ASIN détectés : {len(asins)}")

    results = []

    for i in range(0, len(asins), CHUNK_SIZE):
        chunk = asins[i:i + CHUNK_SIZE]
        print(f"⏳ Traitement chunk {i//CHUNK_SIZE + 1}")

        chunk_results = await process_chunk(chunk)
        results.extend(chunk_results)

        await asyncio.sleep(random.uniform(3, 6))  # pause anti-blocage

    out_df = pd.DataFrame(results)
    out_df.to_csv(OUTPUT_CSV, index=False)

    print("✅ Traitement terminé")
    print(f"📁 Résultat sauvegardé dans : {OUTPUT_CSV}")

# ==========================
# LANCEMENT (JUPYTER SAFE)
# ==========================
await main()
```
